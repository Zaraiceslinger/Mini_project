```python
import pandas as pd
df= pd.read_csv('/content/Placement_Data.csv')
print(df)
```

```python
num_rows, num_columns = df.shape
print(f"Number of rows: {num_rows}, Number of columns: {num_columns}")
```

```python
df.info()
```

```python
average_ssc_p = df['ssc_p'].mean()

print(f"The average for ssc_p is: {average_ssc_p}")
```

```python
max_ssc_p = df['ssc_p'].max()

print(f'the max for ssc_p is: {max_ssc_p}')
```

```python
max_ssc_p = df['ssc_p'].max()
toppers = df[df['ssc_p'] == max_ssc_p]
num_toppers = len(toppers)
print(f"The topper(s) for ssc_p:")
print(toppers)
print(f"\nNumber of toppers: {num_toppers}")
```

```python
max_ssc_p = df.loc[df['ssc_p'].idxmax()]
status_ssc_p = max_ssc_p['status']
print(status_ssc_p)
```

```python
placement_counts = df['status'].value_counts()
print("Number of students placed and unplaced:")
print(placement_counts)
```

```python
unique_degrees_count = df['degree_t'].nunique()
print(f"Number of unique degrees in the dataset: {unique_degrees_count}")
```

```python

correlation = df['ssc_p'].corr(df['hsc_p'])


print(f"Correlation between 10th and 12th percentages: {correlation}")
```

```python
correlation_matrix = df.corr()


print("Correlation Matrix:")
print(correlation_matrix)
```

```python
variances = df.var()
lowest_variance_column = variances.sort_values(ascending=True).index[1]
df = df.drop(lowest_variance_column, axis=1)
print(df)
```

```python
null_counts = df.isnull().sum()
print("Number of Null Values in Each Column:")
print(null_counts)
```

```python
null_counts_before = df.isnull().sum()
print("Number of Null Values in Each Column (Before Imputation):")
print(null_counts_before)
df_filled = df.fillna(df.mean())
null_counts_after = df_filled.isnull().sum()
print("\nNumber of Null Values in Each Column (After Imputation):")
print(null_counts_after)
```

```python
data = {
      'ssc_p': [75, 80, 85, 90, 95],
          'hsc_p': [70, 75, 80, 85, 90]
          }

df = pd.DataFrame(data)
plt.scatter(df['ssc_p'], df['hsc_p'])
plt.xlabel('10th Percentage')
plt.ylabel('12th Percentage')
plt.title('Scatter Plot between 10th and 12th Percentage')
plt.show()
```

```python
data = {
      'ssc_p': [75, 80, 85, 90, 95, 78, 88, 92, 96, 85],
          'hsc_p': [70, 75, 80, 85, 90, 72, 82, 88, 94, 80],
              'status': ['Placed', 'Placed', 'Not Placed', 'Placed', 'Not Placed',
                             'Placed', 'Placed', 'Not Placed', 'Placed', 'Not Placed']
                             }

df = pd.DataFrame(data)
sns.scatterplot(data=df, x='ssc_p', y='hsc_p', hue='status', palette='Set1')

plt.xlabel('10th Percentage')
plt.ylabel('12th Percentage')
plt.title('Scatter Plot between 10th and 12th Percentage (Grouped by Placement Status)')

plt.legend(title='Placement Status')
plt.show()
```

```python
data = {
      'hsc_p': [70, 75, 80, 85, 90, 72, 82, 88, 94, 80],
          'status': ['Placed', 'Placed', 'Not Placed', 'Placed', 'Not Placed',
                         'Placed', 'Placed', 'Not Placed', 'Placed', 'Not Placed']
                         }

df = pd.DataFrame(data)
sns.boxplot(x='status', y='hsc_p', data=df)
plt.xlabel('Placement Status')
plt.ylabel('12th Percentage')
plt.title('Boxplot for 12th Percentage of Students (Grouped by Placement Status)')
plt.show()
```

```python
data = {
      'ssc_p': [75, 80, 85, 90, 95],
          'hsc_p': [70, 75, 80, 85, 90],
              'degree_p': [65, 70, 75, 80, 85],
                  'mba_p': [60, 65, 70, 75, 80]
                  }
df = pd.DataFrame(data)
correlation_matrix = df.corr()
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
data = {
        'ssc_p': [75, 80, 85, 90, 95],
        'hsc_p': [70, 75, 80, 85, 90],
        'degree_p': [65, 70, 75, 80, 85],
        'mba_p': [60, 65, 70, 75, 80]
       }
df = pd.DataFrame(data)
correlation_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Heatmap of Correlation Matrix')
plt.show()
```
